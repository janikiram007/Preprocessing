```
import pandas as pd
import numpy as np
```

```
missing_values = ["?", "none", "NaN"]
dataset = pd.read_csv("credit.csv", na_values = missing_values)
dataset
```

**Handling Missing values**

```
dataset.isna().sum()
```

```
dataset.dropna(axis=0,how='all',thresh=3,inplace=True)
dataset
```

```
dataset.isna().sum()
```

```
dataset.dtypes
```

```
median = dataset['checking_amt'].median()
dataset['checking_amt'].fillna(median, inplace=True)
dataset.isna().sum()
```

```
median = dataset['duration'].median()
dataset['duration'].fillna(median, inplace=True)
dataset.isna().sum()
```

```
dataset['credit_history'].fillna(axis=0, method='ffill', inplace=True)
dataset.isna().sum()
```

```
dataset['purpose'].fillna(axis=0, method='ffill', inplace=True)
dataset.isna().sum()
```

```
median = dataset['credit_amount'].median()
dataset['credit_amount'].fillna(median, inplace=True)
dataset.isna().sum()
```

```
median = dataset['savings'].median()
dataset['savings'].fillna(median, inplace=True)
dataset.isna().sum()
```

```
dataset.dtypes
```

```
dataset['employment'].fillna(axis=0, method='ffill', inplace=True)
dataset.isna().sum()
```

```
median = dataset['installment_commitment'].median()
dataset['installment_commitment'].fillna(median, inplace=True)
dataset.isna().sum()
```

```
dataset['personal_status'].fillna(axis=0, method='ffill', inplace=True)
dataset.isna().sum()
```

90% of the rows for column "other_parties" has missing values, so dropping column "other_parties"

```
dataset.drop('other_parties', axis=1, inplace=True)
```

```
median = dataset['residence_since'].median()
dataset['residence_since'].fillna(median, inplace=True)
dataset.isna().sum()
```

```
dataset['property_magnitude'].fillna(axis=0, method='ffill', inplace=True)
dataset.isna().sum()
```

```
median = dataset['age'].median()
dataset['age'].fillna(median, inplace=True)
dataset.isna().sum()
```

80% of the rows for the column other_payment_plans has missing values so dropping it

```
dataset.drop('other_payment_plans', axis=1, inplace=True)
```

```
dataset['housing'].fillna(axis=0, method='ffill', inplace=True)
dataset.isna().sum()
```

```
median = dataset['existing_credits'].median()
dataset['existing_credits'].fillna(median, inplace=True)
dataset.isna().sum()
```

```
dataset['job'].fillna(axis=0, method='ffill', inplace=True)
dataset.isna().sum()
```

```
median = dataset['num_dependents'].median()
dataset['num_dependents'].fillna(median, inplace=True)
dataset.isna().sum()
```

more than 50% of the rows for the column "own_telephone" has missing values, so dropping it

```
dataset.drop('own_telephone',axis=1,inplace=True)
```

```
dataset.drop('verified',axis=1,inplace=True)
```

```
dataset.dropna(axis=0,how='any',inplace=True)
```

```
dataset.isna().sum()
```

*All the missing values are handled*

```
df=dataset
df
```

```
dataset = df
```

```

dataset['class'] = dataset['class'].str.replace('(?i)bad','0')
dataset['class'] = dataset['class'].str.replace('(?i)b','0')
dataset['class'] = dataset['class'].str.replace('(?i)Good','1')
dataset['class'] = dataset['class'].str.replace('(?i)G','1')



dataset
```

```

dataset['foreign_worker'] = dataset['foreign_worker'].str.replace('r(?i)no','0')
dataset['foreign_worker'] = dataset['foreign_worker'].str.replace('N|n','0')
dataset['foreign_worker'] = dataset['foreign_worker'].str.replace('YES|yes','1')
dataset['foreign_worker'] = dataset['foreign_worker'].str.replace('Y|y','1')
dataset['foreign_worker'] = dataset['foreign_worker'].str.replace(r'(?i)0o','0')


dataset
```

```
dataset['works_outside_US'] = dataset['works_outside_US'].str.replace('r(?i)no','0')
dataset['works_outside_US'] = dataset['works_outside_US'].str.replace('N|n','0')
dataset['works_outside_US'] = dataset['works_outside_US'].str.replace('YES|yes','1')
dataset['works_outside_US'] = dataset['works_outside_US'].str.replace('Y|y','1')
dataset['works_outside_US'] = dataset['works_outside_US'].str.replace(r'(?i)0o','0')
dataset
```

```
dataset.dtypes
```

```
dataset['works_outside_US'] = pd.to_numeric(dataset['works_outside_US'], errors='coerce')
```

```
dataset['class'] = pd.to_numeric(dataset['class'], errors='coerce')           
```

```
dataset['foreign_worker'] = pd.to_numeric(dataset['foreign_worker'], errors='coerce')
```

```
dataset['application_date'] =  pd.to_datetime(dataset['application_date'])
```
